````python
print("\n🎓 Top 10 Colleges:")
print(df['college'].value_counts().head(10))
````

````python
# 6.1 Filter: High scorers (pts > 25)
high_scorers = df[df['pts'] > 25]
print(f"\n⭐ Players averaging >25 points: {len(high_scorers)}")
print(high_scorers[['player_name', 'season', 'pts', 'team_abbreviation']].head(10))
````

````python
#6.2 Filter: Young players (age < 21)
young_players = df[df['age'] < 21]
print(f"\n👶 Players under 21: {len(young_players)}")
print(young_players[['player_name', 'age', 'pts', 'season']].head(10))
````

````python
#6.3 Filter: Specific season (e.g., 2020-21)
season_2021 = df[df['season'] == '2020-21']
print(f"\n📆 2020-21 Season Entries: {len(season_2021)}")
print(season_2021[['player_name', 'team_abbreviation', 'pts']].head(10))
````

````python
#6.4 Filter: Multiple conditions (USA players, age 25-30, pts > 15)
filtered = df[
    (df['country'] == 'USA') & 
    (df['age'] >= 25) & 
    (df['age'] <= 30) & 
    (df['pts'] > 15)
]
print(f"\n🎯 USA Players, Age 25-30, >15 PPG: {len(filtered)}")
print(filtered[['player_name', 'age', 'pts', 'season']].head(10))
````

````python
# Step 7: Grouping & Aggregation
#7.1 Average points by season
avg_pts_by_season = df.groupby('season')['pts'].mean().sort_index()
print("\n📊 Average Points Per Season:")
print(avg_pts_by_season)
````

````python
#7.2 Player with most entries (appeared in most seasons)
player_appearances = df['player_name'].value_counts()
print("\n👤 Players with Most Season Entries:")
print(player_appearances.head(10))
````

````python
#7.3 Average stats by country (top 10)
country_stats = df.groupby('country')[['pts', 'reb', 'ast']].mean().round(2)
print("\n🌍 Average Stats by Country (Top 10 by entries):")
top_countries = df['country'].value_counts().head(10).index
print(country_stats.loc[top_countries])
````

````python
#7.4 Team performance summary
team_summary = df.groupby('team_abbreviation').agg({
    'pts': 'mean',
    'reb': 'mean', 
    'ast': 'mean',
    'player_name': 'count'  # count of player-season entries
}).rename(columns={'player_name': 'entries'}).round(2)

print("\n🏆 Team Summary (Top 10 by avg points):")
print(team_summary.sort_values('pts', ascending=False).head(10))
````

````python
#Step 8: Pattern Discovery
#8.1 Correlation between key stats
print("\n🔗 Correlation Matrix (Key Stats):")
corr_cols = ['age', 'gp', 'pts', 'reb', 'ast', 'net_rating']
print(df[corr_cols].corr().round(3))
````

````python
#8.2 Draft position vs performance
# Only players who were drafted (numeric draft_year)
drafted = df[df['draft_year'].astype(str).str.isdigit()].copy()
drafted['draft_year'] = drafted['draft_year'].astype(int)

print("\n🎲 Avg Points by Draft Round:")
print(drafted.groupby('draft_round')['pts'].mean().round(2))
````

````python
#8.3 Height vs Weight relationship
print("\n📏 Height vs Weight Summary:")
print(df[['player_height', 'player_weight']].describe())

# Simple correlation
print(f"\nHeight-Weight Correlation: {df['player_height'].corr(df['player_weight']):.3f}")
````

````python
# Create the figure
plt.figure(figsize=(10, 6))

# Create histogram
plt.hist(df['pts'].dropna(), bins=30, color='red', edgecolor='black', alpha=0.7)

# Add mean line
plt.axvline(df['pts'].mean(), color='red', linestyle='--', linewidth=2, 
            label=f"Mean: {df['pts'].mean():.1f} PPG")

# Add labels and title
plt.title('Distribution of Points Per Game', fontsize=14, fontweight='bold')
plt.xlabel('Points Per Game')
plt.ylabel('Number of Players')
plt.legend()
plt.grid(axis='y', alpha=0.3)

# Show the plot
plt.tight_layout()
plt.show()
````

````python
# Create the figure
plt.figure(figsize=(10, 6))

# Get top 10 countries
top_countries = df['country'].value_counts().head(10)

# Create horizontal bar chart
plt.barh(top_countries.index, top_countries.values, color='coral', edgecolor='black')

# Add labels and title
plt.title('Top 10 Countries Represented in NBA', fontsize=14, fontweight='bold')
plt.xlabel('Number of Player-Season Entries')
plt.gca().invert_yaxis()  # Put USA at top

# Show the plot
plt.tight_layout()
plt.show()
````

````python
# Create the figure
plt.figure(figsize=(12, 6))

# Calculate average points by season
season_avg = df.groupby('season')['pts'].mean()

# Create line plot
plt.plot(season_avg.index, season_avg.values, marker='o', linewidth=2, color='darkblue')

# Add labels and title
plt.title('Average Points Per Game by Season', fontsize=14, fontweight='bold')
plt.xlabel('Season')
plt.ylabel('Average Points')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels
plt.grid(alpha=0.3)

# Show the plot
plt.tight_layout()
plt.show()
````

````python
# Create the figure
plt.figure(figsize=(10, 6))

# Create scatter plot
plt.scatter(df['player_height'], df['player_weight'], 
            alpha=0.3, s=10, color='purple')

# Add labels and title
plt.title('Height vs Weight Relationship', fontsize=14, fontweight='bold')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.grid(alpha=0.3)

# Show the plot
plt.tight_layout()
plt.show()

# Print correlation
print(f"Height-Weight Correlation: {df['player_height'].corr(df['player_weight']):.3f}")
````

````python
# Create the figure
plt.figure(figsize=(8, 6))

# Create box plot
plt.boxplot(df['age'].dropna(), vert=True, patch_artist=True,
            boxprops=dict(facecolor='lightgreen', color='green'),
            medianprops=dict(color='red', linewidth=2))

# Add labels and title
plt.title('Player Age Distribution', fontsize=14, fontweight='bold')
plt.ylabel('Age (years)')
plt.xticks([1], ['NBA Players'])
plt.grid(axis='y', alpha=0.3)

# Show the plot
plt.tight_layout()
plt.show()

# Print stats
print(f"Median Age: {df['age'].median():.1f} years")
print(f"Age Range: {df['age'].min():.1f} - {df['age'].max():.1f} years")
````

````python
# Create the figure
plt.figure(figsize=(10, 6))

# Get top 10 scorers (minimum 100 games played)
min_games = df[df['gp'] >= 100]
top_scorers = min_games.groupby('player_name')['pts'].mean().nlargest(10)

# Create horizontal bar chart
plt.barh(top_scorers.index, top_scorers.values, color='gold', edgecolor='black')

# Add labels and title
plt.title('Top 10 Players by Average Points Per Game', fontsize=14, fontweight='bold')
plt.xlabel('Average Points Per Game')
plt.gca().invert_yaxis()  # Highest scorer at top

# Show the plot
plt.tight_layout()
plt.show()
````

````python
# Create the figure
plt.figure(figsize=(8, 6))

# Filter drafted players only
drafted = df[df['draft_year'] != 'Undrafted'].copy()
drafted['draft_round'] = pd.to_numeric(drafted['draft_round'], errors='coerce')

# Calculate average points by draft round
round_avg = drafted.groupby('draft_round')['pts'].mean()

# Create bar chart
plt.bar(round_avg.index, round_avg.values, color='red', edgecolor='black')

# Add labels and title
plt.title('Average Points by Draft Round', fontsize=14, fontweight='bold')
plt.xlabel('Draft Round')
plt.ylabel('Average Points Per Game')
plt.xticks([1, 2], ['Round 1', 'Round 2'])
plt.grid(axis='y', alpha=0.3)

# Show the plot
plt.tight_layout()
#plt.savefig('chart_name.png', dpi=300, bbox_inches='tight')
plt.show()

# Print comparison
print(f"Round 1 Avg: {round_avg.get(1, 0):.2f} PPG")
print(f"Round 2 Avg: {round_avg.get(2, 0):.2f} PPG")
````
